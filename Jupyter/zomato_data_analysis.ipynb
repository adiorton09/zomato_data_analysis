# Import the data source file 

```python
import pandas as pd
df1= pd.read_csv("zomato.csv")
```

# Dropping the unnecessary columns.

```python
df2=df1.drop(columns=['url','phone','rest_type','dish_liked','reviews_list','menu_item','listed_in(city)'] , axis=1)
```

# Renaming certain columns for easier reading

```python
df3=df2.rename(columns={'approx_cost(for two people)':'two_people_cost','listed_in(type)':'type_of_restaurant'})

```

```python
df3
```

# Renaming rate to rating

```python
df4=df3.rename(columns={'rate':'rating'})
```

```python
df4
```

# Dropping null value rows on the basis of location , cuisines, two_people_cost

```python
df4=df4.dropna(subset=['location'])
df4=df4.dropna(subset=['cuisines','two_people_cost'])
```

```python
df4
```

# Converting two_people_cost into int data type, then diving it by 2 to get cost for 1 person and renaming it to cost for one

```python
df4['two_people_cost']=df4['two_people_cost'].str.replace(',', '').astype(int)
df4['two_people_cost']=df4['two_people_cost']/2
df4=df4.rename(columns={'two_people_cost':'cost_per_person'})
```

```python
df4
```

# Getting info on the remaining null values

```python
df4.isna().sum()
```

```python
df4['rating']
```

# Remove any non-numeric characters and convert to float

```python
def handlerate(x):
    try:
        return float(x.split('/')[0])
    except:
        return None
```

# Applying the function to modify the rating to a float

```python
df4['rating'] = df4['rating'].apply(handlerate)
```

```python
df4
```

```python
df4.info()
```

# Filling null values with the mean of ratings.

```python
df4['rating'] = df4['rating'].fillna(df4['rating'].mean)
```

```python
df4.isna().sum()
```

# Saving cleaned data

```python
df4.to_csv('zomato_cleaned_data_analysis.csv')
```
